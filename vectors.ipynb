```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define your vectors
vectors = np.array([[0, 0, 0, 1, 2, 3],
                    [0, 0, 0, -2, -1, 1],
                    [0, 0, 0, 3, -2, 1]])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot each vector
for vector in vectors:
    ax.quiver(*vector, arrow_length_ratio=0.1)

# Setting the limits of the plot
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])

# Adding labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show the plot
plt.show()
```

```python
    #vectors is a NumPy array where each row represents a vector. The format is [x_origin, y_origin, z_origin, x_direction, y_direction, z_direction].

    #ax.quiver is used to plot the vectors in 3D space. The *vector unpacks each row of the vectors array into the quiver function.

    #arrow_length_ratio is used to control the appearance of the arrowheads.

    #The set_xlim, set_ylim, and set_zlim methods are used to define the limits of the axes.

    #Finally, plt.show() displays the plot.
```
