```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define vectors: These are the vectors you want to visualize. 
# Each vector is represented as an array of 3 elements [x, y, z].
vectors = np.array([[1, 3, 5]])

# Define the center of the 3D space: 
# This is the point from which all vectors will originate.
center = np.array([5, 5, 5])

# Create a 3D plot: This initializes a 3D plotting context.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot each vector starting from the center:
# The loop goes through each vector, plotting it starting from the 'center'.
# 'quiver' is used for plotting vectors in 3D.
for v in vectors:
    ax.quiver(center[0], center[1], center[2], v[0], v[1], v[2], arrow_length_ratio=0.2)

# Setting the limits of axes to center the plot:
# These limits determine the range of values shown on each axis, 
# ensuring that the plot is centered around the defined 'center'.
axis_limit = 10
ax.set_xlim([center[0] - axis_limit/2, center[0] + axis_limit/2])
ax.set_ylim([center[1] - axis_limit/2, center[1] + axis_limit/2])
ax.set_zlim([center[2] - axis_limit/2, center[2] + axis_limit/2])

# Labeling the axes: This adds labels to the X, Y, and Z axes.
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Add a title to the plot for better readability.
plt.title('3D Vector Visualization from Center')

# Display the plot: This renders the plot on the screen.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define two 2D vectors
vector1 = np.array([1, 1])
vector2 = np.array([1, -2])

# Add vectors: This performs element-wise addition of the two vectors.
resultant_vector = vector1 + vector2

# Define the center of the plot
center = np.array([0, 0])

# Create a 2D plot
fig, ax = plt.subplots()

# Plotting the vectors:
# The 'quiver' function plots vectors. The first two arguments are the starting point,
# and the next two are the vector components.

# Plotting the first vector in blue
ax.quiver(center[0], center[1], vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='blue')

# Plotting the second vector in red
ax.quiver(center[0], center[1], vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='red')

# Plotting the resultant vector in green
# The resultant vector is plotted from the same starting point.
ax.quiver(center[0], center[1], resultant_vector[0], resultant_vector[1], angles='xy', scale_units='xy', scale=1, color='green')

# Setting the limits of axes
# The limits are set to encompass the largest values of the vectors
limit = 10
ax.set_xlim([center[0] - limit/2, center[0] + limit/2])
ax.set_ylim([center[1] - limit/2, center[1] + limit/2])

# Setting the grid ticks
# This sets the ticks at every integer value.
ax.set_xticks(np.arange(center[0] - limit/2, center[0] + limit/2 + 1, 1))
ax.set_yticks(np.arange(center[1] - limit/2, center[1] + limit/2 + 1, 1))

# Labeling the axes
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

plt.title('2D Vector Addition Visualization')
plt.grid()
plt.show()
```
